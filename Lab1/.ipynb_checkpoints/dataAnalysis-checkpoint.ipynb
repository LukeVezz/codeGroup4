First, load the various packages and import the data from the file

```python
import numpy as np
import matplotlib.pyplot as plt


def ecs(l, eps1, eps2, epsm, R):
    """
    l : float
        Wavelength
    eps1 : float
        Real part of the gold dielectric function.
    eps2 : float
        Imaginary part of the gold dielectric function.
    epsm : float
        Dielectric constant of the medium.
    R : float
        Radius of the nanoparticle.
    """
    return 24 * np.pi**2 * np.power(R, 3) * epsm**1.5 * eps2 / (l * ((eps1 + 2 * epsm)**2 + eps2**2))

# Read the bulk dielectric data from the files
with open('./data/bulkAu.txt', 'r') as file_bulk:
    # Skip the header and split each line by whitespace, then convert to float if possible
    bulk_data = []
    for line in file_bulk.readlines()[1:]:
        parts = line.strip().split()
        if len(parts) == 3:
            try:
                bulk_data.append([float(parts[0])*1e-9, float(parts[1]), float(parts[2])])
            except ValueError:
                continue
    bulk_data = np.array(bulk_data)

# Read the nanoparticle absorbance data from the files
with open('./data/G4_np.txt', 'r') as file_np:
    # Skip the header and split each line by whitespace, then convert to float if possible
    np_data = []
    for line in file_np.readlines()[18:]:
        parts = line.strip().split()
        if len(parts) == 2:
            try:
                np_data.append([float(parts[0]), float(parts[1])])
            except ValueError:
                continue
    np_data = np.array(np_data)
    np_data[:, 0] *= 1e-9
    np_data = np_data[::-1]

# Read the preAU absorbance data from the files
with open('./data/G4_preAu.txt', 'r') as file_pre:
    # Skip the header and split each line by whitespace, then convert to float if possible
    pre_data = []
    for line in file_pre.readlines()[18:]:
        parts = line.strip().split()
        if len(parts) == 2:
            try:
                pre_data.append([float(parts[0]), float(parts[1])])
            except ValueError:
                continue
    pre_data = np.array(pre_data)

# Plot data for nanoparticle and preAu with improved visualization
plt.figure(figsize=(10,6))
plt.plot(np_data[:,0], np_data[:,1], label='Nanoparticle', color='tab:blue', linewidth=2)
#plt.plot(pre_data[:,0], pre_data[:,1], label='preAu', color='tab:orange', linewidth=2)
plt.xlabel('Wavelength (nm)', fontsize=14)
plt.ylabel('Absorbance', fontsize=14)
plt.title('Absorbance Spectra', fontsize=16)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python

# bulk_data extraction
filtered_bulk = bulk_data[(bulk_data[:, 0] >= 350E-9) & (bulk_data[:, 0] <= 800E-9)]
l = filtered_bulk[:, 0]
eps1 = filtered_bulk[:, 1]
eps2 = filtered_bulk[:, 2]

# Plot extinction cross section
plt.figure(figsize=(10,6))
plt.plot(l, ecs(l, eps1, eps2, 2, 5e-9), label='ECS (theory)', color='tab:green', linewidth=2)
plt.xlabel('Wavelength (nm)', fontsize=14)
plt.ylabel('ECS', fontsize=14)
plt.title('Extinction Cross Section vs Wavelength', fontsize=16)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
steps = 100
R = np.linspace(1e-10, 100e-9, steps)
Rho = np.linspace(1e5, 1e15, steps)
N = np.linspace(1.2, 1.8, steps)

# bulk_data filtering
wl_begin = 420e-9
wl_end = 580e-9
filtered_bulk = bulk_data[(bulk_data[:, 0] >= wl_begin) & (bulk_data[:, 0] <= wl_end)]
l = filtered_bulk[:, 0]
eps1 = filtered_bulk[:, 1]
eps2 = filtered_bulk[:, 2]

# np_data filtering
filtered_np = np_data[(np_data[:, 0] >= wl_begin) & (np_data[:, 0] <= wl_end)]
np_abs = filtered_np[:, 1]

R_, Rho_, N_ = np.meshgrid(R, Rho, N, indexing='ij')
R_ = R_[..., np.newaxis]
Rho_ = Rho_[..., np.newaxis]
N_ = N_[..., np.newaxis]

l_ = l[np.newaxis, np.newaxis, np.newaxis, :]
eps1_ = eps1[np.newaxis, np.newaxis, np.newaxis, :]
eps2_ = eps2[np.newaxis, np.newaxis, np.newaxis, :]

# Absorption calculation. Ordered as (R, Rho, N, l)
absorption = ecs(l_, eps1_, eps2_, N_**2, R_) * np.log10(np.e) * 0.01 * Rho_
dist = absorption - np_abs[np.newaxis, np.newaxis, np.newaxis, :]
chi2 = np.sum(dist**2 / np_abs[np.newaxis, np.newaxis, np.newaxis, :], axis=3)
print("Minimum chi2:", np.min(chi2))

# Find the index of the minimum chi2 value
min_index = np.unravel_index(np.argmin(chi2), chi2.shape)
print("Minimum chi2 index:", min_index)
print("Best R:", R_[min_index])
print("Best Rho:", Rho_[min_index])
print("Best N:", N_[min_index])

# draw best fit
R_best = R_[min_index]
Rho_best = Rho_[min_index]
N_best = N_[min_index]
absorption_best = absorption[min_index]
plt.figure(figsize=(10,6))
plt.plot(l, absorption_best, label='Best Fit', color='tab:red', linewidth=2)
plt.plot(l, np_abs, label='Experimental', color='tab:blue', linewidth=2)
plt.xlabel('Wavelength (nm)', fontsize=14)
plt.ylabel('Absorbance', fontsize=14)
plt.title('Best Fit Absorbance Spectrum', fontsize=16)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Rho vs R
plt.figure(figsize=(8, 6))
levels = 100
c1 = plt.contourf(Rho, R, chi2[:, :, min_index[2]].T, levels=levels, cmap='hot')
plt.colorbar(c1, label='Chi2')
plt.xlabel('Volumetric Density (m^-3)', fontsize=14)
plt.ylabel('Radius (m)', fontsize=14)
plt.title('Chi2 Contour Plot (Rho vs R)', fontsize=16)
plt.scatter(Rho_best, R_best, color='red', label='Best Fit')
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# N vs R
plt.figure(figsize=(8, 6))
c2 = plt.contourf(N, R, chi2[:, min_index[1], :].T, levels=levels, cmap='hot')
plt.colorbar(c2, label='Chi2')
plt.xlabel('Refractive Index (N)', fontsize=14)
plt.ylabel('Radius (m)', fontsize=14)
plt.title('Chi2 Contour Plot (N vs R)', fontsize=16)
plt.scatter(N_best, R_best, color='red', label='Best Fit')
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# N vs Rho
plt.figure(figsize=(8, 6))
c3 = plt.contourf(N, Rho, chi2[min_index[0], :, :], levels=levels, cmap='hot')
plt.colorbar(c3, label='Chi2')
plt.xlabel('Refractive Index (N)', fontsize=14)
plt.ylabel('Volumetric Density (m^-3)', fontsize=14)
plt.title('Chi2 Contour Plot (N vs Rho)', fontsize=16)
plt.scatter(N_best, Rho_best, color='red', label='Best Fit')
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from skimage import measure
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

# --- INPUT ---
chi2_target = 10.0
tolerance = 0.5  # Accept values in [chi2 - tol, chi2 + tol]
chi2_min, chi2_max = chi2_target - tolerance, chi2_target + tolerance

# Mask values within tolerance band
mask = (chi2 >= chi2_min) & (chi2 <= chi2_max)
chi2_masked = np.where(mask, chi2, chi2_target + 10 * tolerance)  # Push others far away

# marching_cubes needs a *continuous surface*, so ensure transition across isosurface exists
verts, faces, _, _ = measure.marching_cubes(
    chi2_masked, level=chi2_target,
    spacing=(R[1] - R[0], Rho[1] - Rho[0], N[1] - N[0])
)

# --- PLOT ---
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d', proj_type='ortho')
mesh = Poly3DCollection(verts[faces], alpha=0.6)
mesh.set_facecolor('cyan')
ax.add_collection3d(mesh)

ax.set_xlabel('R (m)')
ax.set_ylabel('Rho (m)')
ax.set_zlabel('N')
ax.set_title(f'Isosurface of chi  {chi2_target}  {tolerance}')

ax.set_xlim(0, R[-1] - R[0])
ax.set_ylim(0, Rho[-1] - Rho[0])
ax.set_zlim(0, N[-1] - N[0])
plt.tight_layout()
plt.show()
```
